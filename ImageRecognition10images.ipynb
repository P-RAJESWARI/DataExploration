```python
import os
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
# Path to the parent folder containing subfolders with images
parent_folder = 'Dataset'

# Define the subfolder names
subfolders = ['n07715103', 'n07711080', 'n07747607']
annotations={'n07715103':'Cauliflower','n07711080':'French Fries','n07747607':'Orange'}

# List to store all image paths from each subfolder
all_images = []

# Iterate through each subfolder and collect all image paths
for subfolder in subfolders:
    subfolder_path = os.path.join(parent_folder, subfolder)
    if os.path.isdir(subfolder_path):
        # Get the list of image names in the subfolder
        image_names = os.listdir(subfolder_path)
        # Collect all image paths
        all_images.extend([os.path.join(subfolder, img) for img in image_names])

# Shuffle the list of all image paths
random.shuffle(all_images)

# Select the first 10 shuffled images
sampled_images = all_images[:10]

# Iterate through the sampled images, display them, and rename them based on their class
for image_path in sampled_images:
    full_image_path = os.path.join(parent_folder, image_path)
    # Extract the class (annotation) from the image path
    image_class = image_path.split('/')[0]
    new_image_name = f"{image_class}_{os.path.basename(image_path)}"
    sample=image_class.split('\\')
    label_name=sample[0]
    
     # Add text to the image indicating the class name
    cv2.putText(image, annotations[label_name], (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    
    image = cv2.imread(full_image_path)
    
    # Display the image
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(annotations[label_name])
    plt.axis('off')
    plt.show()
    
    
    
    print(f"Image: {image_path}")
    print(f"Class: {image_class}")
    print(f"Class Name: {annotations[label_name]}")
    
    # Rename the image file
    os.rename(full_image_path, os.path.join(parent_folder, new_image_name))
```
